#Elétron lançado numa região de campo elétrico uniforme

###Vamos resolver o problema com base no lançamento vertical da partícula:

Eixo y: movimento variado

  $y(t)=y_0+voy.t-1/2at²$


Eixo x: movimento uniforme

  $x(t)=xo+vox.t$

###1. Obter a expressão para as componentes $a_x$ e $a_y$ da aceleração.





####O módulo da Força Elétrica Resultante sob a ação do Campo Elétrico (na direção +y), a qual aponta para baixo, é dado por:

$F=E*q$  em Newtons

Onde, $q=-e=1.602 * 10⁻19$C

e, $E_y=3.50 * 10³$N/C

####Aplicando a 2° Lei de Newton para encontrar ax e ay, temos:

$F=m*a$

onde, $m=9,11 * 10^-31$kg

logo, $a_y=E_y*q/m$

e, $a_x=0$ pois $v_0$ é constante no eixo x.



###2. Escrever as expressões para cada coordenada como função do tempo: a x(t) e y(t)

$x(t)=xo+vox.t$

$y(t)=y0+voy.t-1/2.at²$

###3. Criar a função trajetoria($x_0, y_0, v_0, teta0, tmax, dt, E_y$) que retorna como saída os vetores t, x e y.

####O valor máximo de $y$ ocorre quando sua derivada é igual a zero, e $y_0=0$:

$y'=0=v_0.sen(teta0)-E_y.q.t/m$

então, $t$:

$t=v_0.m.sen(tetao)/E_y.q$


```
import numpy as np

def trajetoria(xo,yo,vo,tetao,tmax,dt,Ey,m,q):

  vox=vo*np.cos(tetao)
  voy=vo*np.sin(tetao)
  t=np.arange(0.0,tmax,dt)
  x=np.zeros(len(t))
  y=np.zeros(len(t))
  n=t.size

  for i in range(n):

    x[i]=xo+vox*t[i]

    y[i]=t[i]*(voy*np.sin(tetao))-((Ey*q*t[i])/(2*m))

  return t,x,y

xo=0
yo=0
vo=5*(10**6)
tetao=45
dt=0.0002
Ey=3.5*10**3
m=9.11*10**-31
q=1.602*10**-19
#calculando o tmax
tmax=vo*np.sin(tetao)*m/(Ey*q)


t,x,y=trajetoria(xo,yo,vo,tetao,tmax,dt,Ey,m,q)

for i in range(len(t)):

  print(f'{t[i]:.2f}, {x[i]:.2f}, {y[i]:.2f}')
```

###4. Gráfico da trajetória do elétron

```
import matplotlib.pyplot as plt

plt.plot(t,y,"or-")
plt.xlabel("t")
plt.ylabel(r"$y(t)$")
plt.show()
```
